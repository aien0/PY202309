```python
# 할일을 task_list에 추가하는 함수
def add_task(task_list, task):
    task_list.append(task)

# 할일 목록을 출력하는 함수
def display_tasks(task_list):
    if not task_list:
        print("할일 목록이 비어 있습니다.")
    else:
        print("할일 목록: ")
        for i in range(len(task_list)):
            task = task_list[i]
            print(f"{i+1}. {task}")
            
# 할일 목록을 파일에 저장하는 함수
def save_tasks_to_file(task_list, file_name):
    write_fp = open(file_name, "w", encoding="utf8")
    write_fp.writelines(' \n'.join(task_list))
    write_fp.close()
    
# 파일에서 할일 목록을 읽어오는 함수    
def load_tasks_from_file(file_name):
    task_list = []
    read_fp = open(file_name, "r", encoding="utf8")
    lines = read_fp.readlines()
    
    # 파일에서 읽어온 각 줄의 앞뒤 공백을 제거하여 task_list에 추가
    for line in lines:
        task_list.append(line.strip())
    return task_list

file_name = "tasks.txt"
task_list = load_tasks_from_file(file_name)

while True:
    print("\n할일 관리 프로그램")
    print("1. 할일 추가")
    print("2. 할일 목록 표시")
    print("3. 종료")
    
    # 사용자에게 입력 받음
    choice = input("선택: ")
    
    if choice == '1':
        task = input("할일을 입력하세요: ")
        add_task(task_list, task)
        print("할일이 추가되었습니다.")
    elif choice == '2':
        display_tasks(task_list)
    elif choice == '3':
        # 프로그램 종료 전에 할일 목록을 파일에 저장
        save_tasks_to_file(task_list, file_name)
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```
