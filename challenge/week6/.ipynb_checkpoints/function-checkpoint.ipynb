```python
import random

# 1. 랜덤으로 1부터 45까지의 숫자를 반환하는 함수
def get_number():
    return random.randrange(1,46)

# 2. 로또 번호 6개를 설정하는 함수
def set_lotto_numbers():
    lotto = []
    
    while True:
        num = get_number()
        
        # 중복되지 않는 숫자를 리스트에 추가
        if lotto.count(num) == 0:
            lotto.append(num)
            
        # 6개의 숫자가 모이면 종료
        if len(lotto) >= 6:
            break
            
    return lotto

# 3. 사용자 입력과 로또 번호를 비교하여 일치하는 숫자를 반환하는 함수
def match_numbers(user_inputs, lotto_numbers):
    same_numbers = set(user_inputs) & set(lotto_numbers)
    return same_numbers

# 4. 일치하는 숫자의 개수에 따라 등수를 확인하는 함수
def check_prize(same_numbers):
    if same_numbers == 6:
        print("1등입니다")
    elif same_numbers == 5:
        print("2등입니다")
    elif same_numbers == 4:
        print("3등입니다")
    elif same_numbers == 3:
        print("4등입니다")
    else:
        print("no prize :(")

# 메인 프로그램 시작
num = 0
print("** Start Lotto **")

# 1) 로또 번호 설정
print("1) Set the lotto numbers")
#여기서 set_lotto_numbers 함수 사용
lotto_numbers = set_lotto_numbers()

# 2) 사용자 입력 받기
print("2) Set the user inputs") 
user_inputs = []

while True:
    num = int(input("Enter %d-number: "% (len(user_inputs)+1)))
    if num < 1 or num > 46:
        print("No valid number")
        continue
    # 중복 체크
    if user_inputs.count(num) == 0:
        user_inputs.append(num)
    else:
        print("Same number")
        continue
        
    # 6개의 숫자가 모이면 종료
    if len(user_inputs) >= 6:
        break

# 3) 숫자 비교
print("3) Match the numbers")
same_numbers = match_numbers(user_inputs, lotto_numbers)

# 4) 등수 확인
print("4) Check the prize")
check_prize(same_numbers)

# 5) 리스트 정렬
print("5) Sort the lists")
lotto_numbers.sort()
user_inputs.sort()

# 결과 요약 출력
print("=== Summary ===")
print("Lotto numbers: ", *lotto_numbers)
print("User numbers: ", *user_inputs)
```
